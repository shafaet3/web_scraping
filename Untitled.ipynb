```python
from bs4 import BeautifulSoup
import requests 
import pandas as pd


url = "https://www.programmableweb.com/apis/directory"


npo_apis = {}
api_no = 0
for i in range(735):

    response = requests.get(url)
    data = response.text
    soup = BeautifulSoup(data,'html.parser')
    apis = soup.find_all('tr',{'class':['odd','even']})

    for api in apis:
        api_name = api.find('td',{'class':'views-field-title'}).text.strip('\n')
        api_url = api.find('td',{'class':'views-field-title'}).find('a').get('href').strip('\n')
        acategory = api.find('td',{'class':'views-field-field-article-primary-category'})
        api_category = acategory.text.strip('\n') if acategory else "N/A"
        api_description = api.find('td',{'class':'views-field-field-api-description'}).text.strip('\n')

        api_no+=1
        npo_apis[api_no] = [api_name, api_url, api_category, api_description]

    
    url_tag = soup.find('a',{'title':'Go to next page'})
    print(url_tag)
    if url_tag.get('href'):
        url= 'https://www.programmableweb.com' + url_tag.get('href')
        print(url)
    else:
        break
        
npo_apis_df = pd.DataFrame.from_dict(npo_apis, orient = 'index', columns = ['Api Name','URL','Category', 'Description'])


npo_apis_df.head()


npo_apis_df.to_csv('npo_apis.csv')
```
